```python
m = folium.Map((37.5176577, 126.8864088), zoom_start=16) # tiles='Stamen Watercolor' 이런식으로
# 지도에 위치 표시 하고 싶으면 마커 쓰면 됨다
folium.Marker((37.5176577, 126.8864088), 
              popup='새싹', 
              tooltip='여긴 어디', 
              icon=folium.Icon(icon='cloud')).add_to(m) 
# tooltip은 마우스ㅡ 갖다대면, popup은 클릭하면
folium.CircleMarker((37.5176577, 126.8864088), 
                    radius=20, 
                    color='red', 
                    fill_color='orange', fill_opacity=0.5, 
                    popup='새싹', tooltip='난 누구').add_to(m) # fill_opacity=0.5는 숫자 클수록 진함
m
```
