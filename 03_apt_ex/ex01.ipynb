```python
first_df_melt.columns
# 라스트랑 퍼스트랑 비교해보면 컬럼명이 지역, 지역명으로 다르ㅡㅁ
# 기간 vs 연도, 월 도 다름 <<< 기간을 연도, 월로 나눠야!
```

```python
first_df_melt['기간'].str.split('년').str.get(0)
# 2013년 8월을 2013/ 8월 로 나누고 
# 2013이 돼있는지 확인해봄
```

```python
first_df_melt['기간'].str.split('년').str.get(0).astype('int')
# astype('int') str인 2013을 int 2013으로 변경
```

```python
first_df_melt['연도'] = first_df_melt['기간'].str.split('년').str.get(0).astype('int')
# 2013을 잘라낸 걸 연도 컬럼 만들어서 저장
```

```python
first_df_melt['기간'].str.split('년').str.get(1)
```

```python
first_df_melt['기간'].str.split('년').str.get(1).str.replace('월', '')
```

```python
first_df_melt['월'] = first_df_melt['기간'].str.split('년').str.get(1).str.replace('월', '').astype('int')
```

```python
first_df_melt.columns
```

```python
first_df_melt.columns=['지역명', '기간', '평당분양가격', '연도', '월']
```

```python
cols = ['지역명', '연도', '월', '평당분양가격']
```

```python
last_df
```

```python
# last_df 중에 전용면적 전체인 것만 쓰자
last_df[last_df['전용면적'] =='전체']
```

```python
last_df.loc[last_df['전용면적'] =='전체', cols]
# 전용면적 전체 조건을 만족하는 것중에 cols에 해당하는 것만 불러옴
```

```python
data_last = last_df.loc[last_df['전용면적'] =='전체', cols]
```

```python
data_first = first_df_melt[cols]
# 퍼스트 데이터 중 cols에 해당하는 컬럼만 가져와서 저장
```

```python
result = pd.concat([data_last, data_first])
# 퍼스트 라스트 연결해서 result에 저장
```
