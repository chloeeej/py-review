#### 선형회귀모델 학습 예측 평가

```python

def get_rmse(model):
    from sklearn.metrics import mean_squared_error
    import numpy as np
    pred = model.predict(X_test)
    mse = mean_squared_error(y_test, pred)
    rmse = np.sqrt(mse)
    print(model.__class__.__name__, '로그 변환된 RMSE: ', np.round(rmse, 3)) # 무슨 모델인지 모르니까 클래스 이름 나오게
    return rmse

# 위에 get_rmse(model)에서 받은 결과들을 리스트에. 여러개 돌릴 거면 이거에 넣어서 돌리기
def get_rmses(models):
    rmses = []
    for model in models:
        rmse = get_rmse(model)
        rmses.append(rmse)
    return rmses
        
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso # LinearRegression일반선형회귀, Ridge, Lasso 규제들어간 선형회귀
from sklearn.model_selection import train_test_split
```

```python
# p.382
def get_top_bottom_coef(model, n=10):
    coef = pd.Series(model.coef_, index=X.columns) # 회귀계수값?
    coef_high = coef.sort_values(ascending=False).head(n) # 높은 순으로 정렬해서 n개만 뽑아 보여죠
    coef_low = coef.sort_values(ascending=False).tail(n)
    return coef_high, coef_low
```

```python
get_top_bottom_coef(lr_reg,5)
```

```python
# p.382 3가지 모델에 대한 회귀계수값을 그래프로 찍어줌
def visualize_coefficient(models):
    fig, axs = plt.subplots(figsize=(24,10), nrows=1, ncols=3)
    fig.tight_layout() # 레이아웃.. 좀... 맞춰줌..
    for i_num, model in enumerate(models):  # enumerate가 인덱스랑 값을 리턴하니까 i_num에 인덱스 들어가고./
        coef_high, coef_low = get_top_bottom_coef(model)
        coef_concat = pd.concat([coef_high, coef_low])
        axs[i_num].set_title(model.__class__.__name__+'coefficient', size=25)
        axs[i_num].tick_params(axis='y', direction='in', pad=-120)
        for label in (axs[i_num].get_xticklabels() + axs[i_num].get_yticklabels()):
            label.set_fontsize(22)
        sns.barplot(x=coef_concat.values, y=coef_concat.index, ax=axs[i_num])
```

```python
visualize_coefficient(models)
```

```python
# p.383
from sklearn.model_selection import cross_val_score
```

```python
def get_avg_rmse_cv(models):
    for model in models:
        rmse_list = np.sqrt(-cross_val_score(model, X, y,scoring='neg_mean_squared_error', cv=5))
        rmse_avg = np.mean(rmse_list)
        print(f'{model.__class__.__name__}모델 cv rmse값 리스트: {np.round(rmse_list,3)}')
        print(f'{model.__class__.__name__}모델 cv 평균 rmse값: {np.round(rmse_avg,3)}')
```

```python
get_avg_rmse_cv(models)
```

```python
# p.384
from sklearn.model_selection import GridSearchCV
```

```python
def print_best_params(model, params):
    grid_model = GridSearchCV(model, params, scoring='neg_mean_squared_error', cv=5)
    grid_model.fit(X,y)
    rmse = np.sqrt(-1*grid_model.best_score_)
    print(f'{model.__class__.__name__}모델 cv=5일 때 최적 평균 rmse값: {np.round(rmse,4)}, 최적 alpha값: {grid_model.best_params_}')
    
```

```python
ridsge_params = {
    'alpha': [0.05,0.1,1,5,8,10,12,15,20]
}
print_best_params(ridge_reg, ridsge_params)
```

```python
lasso_params = {
    'alpha': [0.001,0.005,0.008,0.05,0.03,0.1,0.5,1,5,10]
}
print_best_params(lasso_reg, lasso_params)
```

```python
lr_reg = LinearRegression()
lr_reg.fit(X_train, y_train)

ridge_reg = Ridge(alpha=12)
ridge_reg.fit(X_train, y_train)

lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)

models = [lr_reg, ridge_reg, lasso_reg]
get_rmses(models)

visualize_coefficient(models)
```
