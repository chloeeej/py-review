```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
data = pd.read_csv('complt_data_proc.csv', encoding='cp949')
```

```python
data.head()
```

```python
fig = plt.figure()
a1 = fig.add_subplot(1,1,1)
a1.scatter(data['연도'], data['PM10 농도'], alpha=0.5)

# 망...! 
```

```python

data_sido = data.groupby(['연도', '시도명'])['PM10 농도'].mean()
data_sido = data_sido.unstack(level='시도명')
data_sido.columns.name = 'PM10 농도'
plt.figure(figsize=(8,6))
# data_sido = data_sido.plot()
data_sido = plt.plot(data_sido)


data_sido.set_xlabel('연도')
data_sido.set_ylabel('미세먼지농도')
```
