```python
import pandas as pd
```

```python
rate = pd.read_csv('rate.csv', encoding='cp949')
rate.drop(columns='Unnamed: 0', inplace=True)
rate
```

```python
pm10_data = pd.read_csv('pm10_data.csv', encoding='cp949')
pm10_data['미세먼지10'] = pm10_data['미세먼지'] 
pm10_data = pm10_data[['기간', '지역', '미세먼지10']]
pm10_data
```

```python
pm25_data = pd.read_csv('pm25_data.csv', encoding='cp949')
pm25_data['미세먼지25'] = pm25_data['미세먼지'] 
pm25_data = pm25_data[['기간', '지역', '미세먼지25']]
pm25_data
```

```python
rate
```

```python
data = pd.merge(rate, pm10_data, left_on=['기간', '지역'], right_on=['기간', '지역'], how='outer')
data = pd.merge(data, pm25_data, left_on=['기간', '지역'], right_on=['기간', '지역'], how='outer')
data

data.to_csv('data.csv', encoding='cp949')
```
