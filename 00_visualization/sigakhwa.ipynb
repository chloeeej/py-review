```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
data = pd.read_csv('complt_data_proc.csv', encoding='cp949')
```

```python
#  
plt.title('   ', fontsize=14, pad=15)

#  
plt.rcParams['figure.figsize'] = (6, 3)

#  
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')['PM10 '], label='', linestyle='solid')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')['PM10 '], label='', linestyle='dashed')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')['PM10 '], label='', linestyle='dotted')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')['PM10 '], label='', linestyle='dashdot')

#  
plt.grid(True)

#  
plt.xlabel('', fontsize=12, labelpad=15)
plt.ylabel('PM10  (: g/m3)', fontsize=12, labelpad=15)

# ?!  
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

#  
plt.legend(loc=(1.0, 0.0), fontsize=10, frameon=False)
# plt.legend(loc=(1.0, 0.0), ncol=2) #  

plt.show()
```

```python
#  
plt.title('   ', fontsize=14, pad=15)

#  
plt.rcParams['figure.figsize'] = (6, 3)

#  
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')[' ']*10000, label='', linestyle='solid')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')[' ']*10000, label='', linestyle='dashed')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')[' ']*10000, label='', linestyle='dotted')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')[' ']*10000, label='', linestyle='dashdot')

#  
plt.grid(True)

#  
plt.xlabel('', fontsize=12, labelpad=15)
plt.ylabel('  ( )', fontsize=12, labelpad=15)

# ?!  
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

#  
plt.legend(loc=(1.0, 0.0), fontsize=10, frameon=False)

plt.show()
```

```python
# y       

plt.title('   ', fontsize=14, pad=15)
plt.rcParams['figure.figsize'] = (6, 3)

plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')['PM10 '], label='', linestyle='solid')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')['PM10 '], label='', linestyle='dashed')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')['PM10 '], label='', linestyle='dotted')
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')['PM10 '], label='', linestyle='dashdot')
# y       
plt.plot(data.groupby(['']).get_group('')[''], data.groupby(['']).get_group('')[' '], label=' ')
plt.grid(True)
plt.xlabel('', fontsize=12, labelpad=15)
plt.ylabel('PM10  (: g/m3)', fontsize=12, labelpad=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(loc=(1.0, 0.0), fontsize=10, frameon=False)
plt.show()
```

```python
#  

#    
fig, ax1 = plt.subplots()

#  
plt.title('    ', fontsize=14, pad=15)

#  
plt.rcParams['figure.figsize'] = (6, 3)

# x  y     
ax1.plot(data[''].unique(), data.groupby([''])['PM10 '].mean(), label=' PM10 ', color='pink')
ax1.set_xlabel('', fontsize=12, labelpad=15)
ax1.set_ylabel('  (: g/m3)', fontsize=12, labelpad=15)
ax1.legend(loc='upper left', fontsize=10, frameon=False)

# ax1 x   Axes  
ax2 = ax1.twinx()

# y   
ax2.plot(data[''].unique(), data.groupby([''])[' '].mean()*10000, label=' ', color='mediumpurple')
ax2.set_ylabel('  ( )', fontsize=12, labelpad=15)
ax2.legend(loc='lower left', fontsize=10, frameon=False)


#  
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```
