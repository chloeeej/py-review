### 열 이름 관리하기 

#### 하나의 열이 여러 의미를 가지고 있는 경우

```python
df = pd.read_csv('data/country_timeseries.csv')
```

```python
df.sample(5)
```

```python
df_melt = pd.melt(df, id_vars=['Date', 'Day'])
```

```python
df_melt
```

#### split 메서드로 열 이름 분리하기 

```python
# variable을 쪼개서 cases와 Deaths를 나누고, 나라를 나눠야..
# 시리즈에 바로 split 메서드를 쓸 수는 없어서 시리즈를 str으로 바꿔놓고 split 해야

# df_melt.variable 점 찍는 방법. 타수룰 줄일 수 있따 근데 안되는 경우도 있음
# df_melt(... 방법이랑 비교
df_melt.variable
```

```python
df_melt.variable.str.split('_')
# _를 기준으로 나누겠다. 분리를 했으니 리스트 형식으로 나옴
```

```python
df_melt.variable.str.split('_')[0]
# 첫번째 행이 선택됨..
```

```python
df_melt.variable.str.split('_').str.get(0)
# str의 0번을 가져오기
```

```python
df_melt['status'] = df_melt.variable.str.split('_').str.get(0)
df_melt['country'] = df_melt.variable.str.split('_').str.get(1)
# 새로운 컬럼을 만들 떄는 [] 안에 없는 이름 주면 됨
```

```python
df_melt.sample(5)
```

### 중복 데이터 처리하기

```python
df= pd.read_csv('data/billboard.csv')
```

```python
df.sample(5)
```

```python
df.columns
```

```python
df_melt = pd.melt(df, id_vars=['year', 'artist', 'track', 'time', 'date.entered'], 
                  var_name='week', 
                  value_name='rating')
```

```python
df_melt
```

```python
df_melt.shape
```

```python
df.shape
```

```python
df_melt[df_melt.track == 'Kryptonite']
# df_melt의 track 값이 크립토어쩌고인 거. 같은 제목의 값이 몇개나 있냐
```

```python
df_songs = df_melt[['year', 'artist', 'track', 'time', 'date.entered']]
```

```python
df_songs.shape
```

```python
df_songs.drop_duplicates(inplace=True)
# 중복 제거
```

```python
df_songs.shape
# 위에서 드랍_듀플리케이트 하기 전에는 (24092, 5) 였는데 (317, 5)로 줄어듦
```

```python
df_songs['id']=range(len(df_songs))
# id라는 컬럼 만들어주기
```

```python
df_songs.head()
```

```python
# 아이디를 키로 주고 나머지를 제거 뭔말인연ㅇ밓ㅁㄴㅇ;ㅣㅏ 아아아앙아 ㅠㅠ
df_melt.merge(df_songs, on=['year', 'artist', 'track', 'time', 'date.entered'])
```

```python
df_ratings = df_melt.merge(df_songs, on=['year', 'artist', 'track', 'time', 'date.entered'])
```

```python
# 필요 없는 부분 제거하기
df_ratings = df_ratings[['id', 'date.entered', 'week', 'rating']]
```
