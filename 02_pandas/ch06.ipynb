#### 누락값 처리하기 
변경하거나, 삭제하거나

```python
ebola.fillna(0)
# na값을 채워 원본은 안 건드리고 채운 값을 리턴만 해줌
# 평균값에 영향을 안 주고 싶으면 0 말고 평균값을 넣기도 하고
# 누적값인 경우 0으로 채우면 곤란하니까 앞쪽이나 뒤쪽값으로 넣거나 하기도 함 
```

```python
ebola.fillna(method='ffill')
# 앞쪽 값으로 채워졌네여 인덱스 3이 비어있으면 인덱스 2의 값이 들어감
```

```python
ebola.fillna(method='bfill')
# 뒤의 값으로 채워짐
```

```python
ebola.interpolate()
# 앞뒤 중간값으로 채워줌
```

```python
ebola.dropna()
# 누락값이 있는 건 삭제해버리기
```
