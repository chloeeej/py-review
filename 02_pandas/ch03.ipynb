### 나만의 데이터 만들기

시리즈와 데이터프레임 직접 만들기

```python
import pandas as pd
```

시리즈 만들기

```python
s = pd.Series(['banana', 42])
# 클래스명은 대문자로 시작, 함수 이름은 소문자로 시작하는 규칙이 여기서도 통용
```

```python
s
```

```python
s = pd.Series(['Wes McK', 'pandas'], index=['person', 'who'])
# 인덱스값을 직접 지정할 수도 있다
```

```python
s
```

데이터프레임 만들기

```python
scientists = pd.DataFrame({
    'name' : ['R Franklin', 'W Gosset'],
    'occupation' : ['chemist', 'statistician'],
    'born' : ['1920-07-25', '1876-06-13'],
    'age' : [37, 61] })
print(scientists)
```

```python
scientists.info()
```


인덱스랑 컬럼 따로 넣기도 함
컬럼을 따로 넣으면, 컬럼은 리스트고 리스트는 순서가 있으니까 컬럼 순서대로 만들어짐 (딕셔너리는 순서가 내맘대로 안 될 수 있음 ㅠㅠ)
