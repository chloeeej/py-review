```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
fig = plt.figure()
hex2 = fig.add_subplot(1,1,1)
hex2.hexbin(tips['total_bill'], tips['tip'],cmap='PuBu')
```

```python
fig3 = plt.figure()
a1 = fig3.add_subplot(1,1,1)
a1.scatter(tips['total_bill'], tips['tip'], s=tips['size']*10, alpha=0.5)
```

```python
hex = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
aaa= sns.heatmap(tips)
```
