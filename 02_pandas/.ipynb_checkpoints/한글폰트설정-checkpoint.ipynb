```python
import numpy as np
```

```python
# 대충 아무 데이터 만들어보기
data = np.random.randint(-100, 100, 50).cumsum()
# cumsum 누적합계
data
```

```python
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
# 경고메시지 안 뜨게 하기
```

```python
plt.plot(range(50), data)
# x값은 범위로 50까지, y값은 data를
```

```python
plt.plot(range(50), data, 'r')
plt.title('다이어트 현황')
plt.ylabel('체중')
plt.xlabel('날짜')
```

```python
# 한글 폰트  설치해주기
# 맑은고딕 Malgun Gothic
```

```python
import matplotlib.font_manager as fm
```

```python
# 1. 실제 파일을 가지고 하는 법 
# 폰트매니저 임포트
# 폰트 파일을 폴더에 복사해놓고
# font properties 사용
path = 'malgun.ttf' # 파일 경로가 필요
fontprop = fm.FontProperties(fname=path, size=16)
```

```python
plt.plot(range(50), data, 'r')
plt.title('다이어트 현황', fontproperties=fontprop)
plt.ylabel('체중', fontproperties=fontprop)
plt.xlabel('날짜', fontproperties=fontprop)
```

```python
# 2. 전역글꼴 설정 rcParams. 사용할 때 한번 선언 해주고 써야함
# 폰트를 직접 정해주는 건 해당 그래프에만 적용되니까, 세션 도는 동안 전부 적용될 수 있게

plt.rcParams['font.size']
```

```python
plt.rcParams['font.family']
# 기본 글꼴
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.plot(range(50), data, 'r')
plt.title('다이어트 현황')
plt.ylabel('체중')
plt.xlabel('날짜')
```
