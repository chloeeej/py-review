## 데이터 연결하기

### 데이터 연결 기초

분석하기 좋게 만드는 과정을 클렌징 작업이라고 함

```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/concat_1.csv')
df2 = pd.read_csv('data/concat_2.csv')
df3 = pd.read_csv('data/concat_3.csv')
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# 데이터 연결시키기
pd.concat([df1, df2, df3])
# 이너조인 키값이 일치하는 걸로 갖고옴, 아우터조인 키값이 일치하는 게 없어도 일단 가져옴(없는건 null값 넣음). 안 쓰면 아우터조인
# 컬럼이 같은 것끼리 붙음
```

```python
pd.concat([df1, df2, df3], axis=1)
# axis=1 옆으로 붙어여. 인덱스가 같은 것끼리
```

```python
# 데이터프레임에 시리즈 연결하기
new = pd.Series(['n1', 'n2', 'n3', 'n4'])
```

```python
pd.concat([df1, new])
```

```python
pd.DataFrame([['n1', 'n2', 'n3', 'n4']])
```

```python
pd.DataFrame(['n1', 'n2', 'n3', 'n4'])
```

```python
new_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=['A', 'B', 'C', 'D'])
# 컬럼 맞추기
```

```python
pd.concat([df1, new_df])
# 컬럼 맞아서 아래에 가서 붙음
```

#### 다양한 방법으로 데이터 연결하기

```python
pd.concat([df1, df2, df3], axis=0, ignore_index=True)
# 기존 인덱스 무시하고 새로 인덱스 부여함?!
```

### 데이터 연결 마무리

```python
person = pd.read_csv('data/survey_person.csv')
site = pd.read_csv('data/survey_site.csv')
survey = pd.read_csv('data/survey_survey.csv')
visited = pd.read_csv('data/survey_visited.csv')
```

```python
visited
```

```python
sub_visited =visited.loc[[0, 2, 6]]
```

```python
sub_visited
```

```python
site
```

```python
# site랑 visited 같은 값이 있는 거 보니 이거ㅓㄹ 연결하면 되겠구나
site.merge(sub_visited, left_on='name', right_on='site')
# 같은 값의 컬럼명이 같으면 안 써줘도 되는데 다르니까 어느 컬럼 가져올지 써줘야함 left(site)는 뭘로 할거냐 네임. right는 사이트로
```

concat은 컬럼이나 인덱스가 같은 경우  
merge는 같은 데이터값이 있는 경우 (sql의 join이랑 같은 개념?!)
