#### 행단위 데이터 추출하기
iloc 인덱스를 기준으로추춘
loc 행 기준 추출??

````python
df.head(3)
````

````python
df.loc[2]
# 2번 행의 데이터를 가져와라
````

````python
df.loc[2:5]
# 2번부터 5번까지 가져와
````

````python
df.loc[[1,5,10]]
# 1, 5, 10에 있는 거 갖고와라. 하나가 아니니까 하나로 묶어줄 것
````

````python
df.loc[1, 'country']
# 1에 country 정보 가져와
````

````python
df.loc[1:6, 'country':'year']
# 1부터 6까지, 컨트리부터 이어까지
````

````python
df.loc[[1,6], ['country', 'year']]
````

````python
df.iloc[0]
# 첫번째 행에 있는 거
````

````python
df.iloc[0, 0]
# 첫번째 행 첫번째 열에 있는 거
````

````python
df.iloc[0:1,0:3]
````

````python
df.iloc[-1]
# 뒤에서 첫번째거
````

### 기초적인 통계 계산하기

평균 구하기

````python
df.mean(numeric_only=True)
````

````python
df.groupby('year')['lifeExp'].mean()
# year별 lifeExp.평균
# 1952년의 기대수명은 평균 49.05세..
````

````python
df.mean(numeric_only=True,axis=1)
#뭐지.. 행기준 열기준 나오면 모르겄따 ㅠ
````

````python
df.groupby(['year', 'continent'])['lifeExp'].mean()
````

````python
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
````
